```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
diab=datasets.load_diabetes( return_X_y=False, as_frame=False, scaled=True)
```

```python
diab.keys()
```

```python
diab.values()
```

```python
diab['target']
```

```python
df_diab = pd.DataFrame(diab['data'],columns = diab['feature_names'])
df_diab
```

```python
df_diab['target']=diab['target']   #target-> progression of diabetes after one year
df_diab
```

```python
diab_model=linear_model.LinearRegression()
```

```python
X=df_diab[diab['feature_names']]
y=diab['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
diab_model.fit(X_train, y_train)
```

```python
y_pred = diab_model.predict(X_test)
y_pred
```

```python
y_test
```

```python
#to calculate the mean absolute error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```

```python
#root_mean_squared_error
from sklearn.metrics import root_mean_squared_error
print(root_mean_squared_error(y_test, y_pred))
```

```python
#r^2 score
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_percentage_error
print(mean_absolute_percentage_error(y_test, y_pred))
```
